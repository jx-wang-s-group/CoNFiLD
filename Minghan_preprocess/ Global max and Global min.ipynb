# Global max and Global min

## Without Reshaping

```python
import numpy as np

# Load full dataset
data = np.load("/Users/minghan/Library/CloudStorage/GoogleDrive-john.chuicandoit@gmail.com/My Drive/CoNFiLD/latents/Case3/train_data.npy")  # Shape: (time_steps, 10884, 2)

# Compute global max and min over all time steps and spatial points
data_max = np.max(data, axis=(0, 1))  # Get max for each variable (u, v)
data_min = np.min(data, axis=(0, 1))  # Get min for each variable (u, v)

print("Max values:", data_max)  # Should be (2,) → max for (u, v)
print("Min values:", data_min)  # Should be (2,) → min for (u, v)
```

```python
import numpy as np

# Load latent representations (NOT raw velocity data!)
train_latents = np.load("/Users/minghan/Library/CloudStorage/GoogleDrive-john.chuicandoit@gmail.com/My Drive/CoNFiLD/latents/Case3/train_data.npy")   # Shape: (1692, 256, 256)

# Extract max and min along the training samples dimension (axis=0)
max_val = np.max(train_latents, axis=0)  # Shape: (256, 256)
min_val = np.min(train_latents, axis=0)  # Shape: (256, 256)

# Save in correct shape
np.save("./max_min/max_val.npy", max_val)
np.save("./max_min/min_val.npy", min_val)

print(f"[INFO] Saved max_val shape: {max_val.shape}")
print(f"[INFO] Saved min_val shape: {min_val.shape}")
```
