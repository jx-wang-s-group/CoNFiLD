```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os

# Enable interactive mode for rotation
plt.ion()

# Base directory for sensor folders
base_dir = "/Users/minghan/Library/CloudStorage/GoogleDrive-john.chuicandoit@gmail.com/My Drive/CoNFiLD/ConditionalDiffusionGeneration/inference_scripts/Case4/random_sensor/input/random_sensor/"

# Sensor folders (10, 100, 1000)
sensor_folders = ["10", "100", "1000"]

# Define colors for each case
colors = ["red", "blue", "green"]

# Define marker size (adjustable flag)
marker_size = 5  # Change this to adjust point size

# Create a figure with subplots
fig = plt.figure(figsize=(15, 5))
num_subplots = len(sensor_folders)

# Create results folder if it doesn't exist
results_dir = "results"
os.makedirs(results_dir, exist_ok=True)

for i, folder in enumerate(sensor_folders):
    coords_path = os.path.join(base_dir, folder, "coords.npy")
    
    if not os.path.exists(coords_path):
        print(f"Skipping {folder}: File not found -> {coords_path}")
        continue
    
    # Load sensor coordinates
    coords = np.load(coords_path)
    x, y, z = coords[:, 0], coords[:, 1], coords[:, 2]

    # Create subfigure
    ax = fig.add_subplot(1, num_subplots, i + 1, projection='3d')
    ax.scatter(x, y, z, s=marker_size, alpha=0.7, marker="o", color=colors[i])

    # Labels and title
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title(f"Sensor Config: {folder}")

    # Allow rotation
    ax.view_init(elev=30, azim=45)  # Default view (adjustable)

save_path = os.path.join(results_dir, "3D_Sensor_Visualization.png")
plt.savefig(save_path, dpi=300)
print(f"Plot saved to {save_path}")

# Adjust layout and show
plt.tight_layout()
plt.show()

# Keep the plot interactive
plt.ioff()
plt.show()
```
